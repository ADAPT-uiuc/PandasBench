```python
import os

def scale_input_data(scale_factor):
  file_bases = ['./input/train_users_2', './input/test_users']
  for file_base in file_bases:
    import pandas as pd
    import shutil
    if scale_factor == 1.0:
      shutil.copyfile(file_base + '.csv', file_base + '.scaled.csv')
      continue
    df_to_scale = pd.read_csv(file_base + '.csv')
    new_num_rows = int(scale_factor * len(df_to_scale))
    if scale_factor <= 1.0:
      df_to_scale = df_to_scale.iloc[:new_num_rows]
    else:
      while len(df_to_scale) < new_num_rows:
        df_to_scale = pd.concat([df_to_scale, df_to_scale[:min(new_num_rows - len(df_to_scale), len(df_to_scale))]])
    df_to_scale.to_csv(file_base + '.scaled.csv', index=False)

if 'INPUT_SCALE_FACTOR' in os.environ:
  scale_input_data(float(os.environ['INPUT_SCALE_FACTOR']))
```

```python
df_train = pd.read_csv("./input/train_users_2.scaled.csv")
df_train.sample(n=5) #ou head pour avoir les 5 premiers
```

```python
#Combine into one dataset
df_all = pd.concat((df_train, df_test), axis=0, ignore_index=True)
df_all.head(n=5) #only display a few lines and not the whole dataframe
```

```python
df_all.drop('date_first_booking',axis=1, inplace=True)
```

```python
df_all.sample(n=5)
```

```python
df_all['timestamp_first_active'] = pd.to_datetime(df_all['timestamp_first_active'], format='%Y%m%d%H%M%S')
df_all.sample(n=5)
```

```python
def remove_age_outliers(x, min_value=15, max_value=90):
    if np.logical_or(x<=min_value, x>=max_value):
        return np.nan
    else:
        return x
```

```python
df_all['age'] = df_all['age'].apply(lambda x: remove_age_outliers(x) if(not np.isnan(x)) else x)
```

```python
df_all['age'].fillna(-1, inplace=True)
```

```python
df_all.sample(n=5)
```

```python
df_all.age = df_all.age.astype(int)
df_all.sample(n=5)
```

```python
def check_NaN_Values_in_df(df):
    for col in df:
        nan_count = df[col].isnull().sum()
        
        if nan_count != 0:
            print(col + "=>" + str(nan_count) + "NaN Values")
```

```python
check_NaN_Values_in_df(df_all)
```
