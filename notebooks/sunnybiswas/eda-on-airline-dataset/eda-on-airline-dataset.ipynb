```python
import os

def scale_input_data(scale_factor):
  file_bases = ['./input/Airline Dataset']
  for file_base in file_bases:
    import pandas as pd
    import shutil
    if scale_factor == 1.0:
      shutil.copyfile(file_base + '.csv', file_base + '.scaled.csv')
      continue
    df_to_scale = pd.read_csv(file_base + '.csv')
    new_num_rows = int(scale_factor * len(df_to_scale))
    if scale_factor <= 1.0:
      df_to_scale = df_to_scale.iloc[:new_num_rows]
    else:
      while len(df_to_scale) < new_num_rows:
        df_to_scale = pd.concat([df_to_scale, df_to_scale[:min(new_num_rows - len(df_to_scale), len(df_to_scale))]])
    df_to_scale.to_csv(file_base + '.scaled.csv', index=False)

if 'INPUT_SCALE_FACTOR' in os.environ:
  scale_input_data(float(os.environ['INPUT_SCALE_FACTOR']))
```

```python
# Removing the Spaces on the column names and replacing it with (-) 
# Refrencing the dataframe to another variable (df2)

df.columns = df.columns.str.replace(' ', '_')
df2 = df
df2

# Quering on the data based on Airport Country Code (United States) 
US_data = df2.query('Airport_Country_Code == "US"')


US_data
```

```python
#  The Departure Date is an Object, converting it into Date-time format and extracting Quarter, month and Week 

US_data['Date'] = pd.to_datetime(US_data['Departure_Date']) 

# Extract quarter, month, week

US_data['Quarter'] = US_data['Date'].dt.quarter
US_data['Month'] = US_data['Date'].dt.month
US_data['Week'] = US_data['Date'].dt.isocalendar().week
US_data

#  So Now we have a data_set that only has values for United States for further EDA
```

```python
#  Creating a Customer Column based on Departure and Arrival Airport 
US_data['Route'] = US_data['Airport_Name'].str.cat(US_data['Arrival_Airport'], sep=' -to- ')

US_Flight_Route = US_data['Route'].value_counts()
US_Flight_Route = pd.DataFrame({'Route': US_Flight_Route.index, 'Count': US_Flight_Route.values})
Top_20_US_Flight_Route = US_Flight_Route.head(20)

# Plotting a Donut Chart that represents the travelers ratio 

# FIRST-AUTHOR: remove plotting
# fig5 = px.bar(Top_20_US_Flight_Route, y='Route', x='Count')
# fig5.update_layout(title="Ratio of USA Visitors")
# fig5.show()
```

```python
# Figuring out the total number of flight per month 


filtered_df = df[['Airport_Name','Departure_Date','Arrival_Airport','Pilot_Name']]
filtered_df1 = pd.DataFrame(filtered_df)
filtered_df1_no_duplicates = filtered_df1.drop_duplicates()
filtered_df1_no_duplicates['Date'] = pd.to_datetime(filtered_df1_no_duplicates['Departure_Date']) 

# Extracting month
filtered_df1_no_duplicates['Month'] = filtered_df1_no_duplicates['Date'].dt.month
filtered_df1_no_duplicates
```

```python
# Table for Total number of flight per month (This included all flight status)

filtered_df1_no_duplicates
Total_flights = filtered_df1_no_duplicates['Month'].value_counts()
Total_flights = Total_flights.reset_index()
Total_flights.sort_values(by='index', ascending=True, inplace=True)
Total_flights
```
